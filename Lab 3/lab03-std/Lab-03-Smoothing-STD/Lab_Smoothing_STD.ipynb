```python
import math
import commonfunctions as cf # this a custom module found the commonfunctions.py
import skimage.io as io
import matplotlib.pyplot as plt
import numpy as np
from skimage.color import rgb2gray, rgba2rgb
from skimage.filters import median , gaussian as gaussian_filter
# from scipy import fftpack
# from scipy.signal import convolve2d
from skimage.util import random_noise
# Show the matlpotlib figures inside the notebook
%matplotlib inline
%load_ext autoreload
%autoreload 2
```

```python
'''
(1) Your own median filter algorithm
'''
def median_filter(img , size = 3):
    w , h = img.shape
    if w < size or h < size:
        return img
    return_img = np.copy(img)
    for i in range(0 , w - size + 1):
        for j in range(0 , h - size + 1):
            return_img[i+size//2,j+size//2] = np.median(img[i:i+size , j:j+size], axis=None)
    return return_img
img = rgb2gray(rgba2rgb(io.imread('./imgs/Picture1.png')))
noisy_img = random_noise(img, 's&p' , amount=0.05)

applied_median_img = median_filter(noisy_img,3)
cf.show_images([img , noisy_img , applied_median_img] , ['original' , 'noisy' , 'median'])

```

```python
'''
(2) skimage median filter
'''
skiimge_filter = median(img)
cf.show_images([img , noisy_img , applied_median_img , skiimge_filter] , ['original' , 'noisy' , 'median' , 'skiimge filter'])
```

```python
'''
(3) Gaussion Filters
'''
gaussian_img_8 = gaussian_filter(img , sigma=8)
gaussian_img_3 = gaussian_filter(img , sigma=3)
gaussian_img_2 = gaussian_filter(img , sigma=0.02)
cf.show_images([img , noisy_img , gaussian_img_8,gaussian_img_3,gaussian_img_2] , ['original' , 'noisy' , 'gaussian_img 8' , 'gaussian_img 3' , 'gaussian_img 0.2' ])
```
